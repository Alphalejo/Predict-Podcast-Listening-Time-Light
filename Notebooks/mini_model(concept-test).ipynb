```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import pickle
```

```python
encoded_df = pd.read_csv('../Data/processed/encoded_df.csv')
mini_df = encoded_df.sample(n=5000, random_state=42)
X_mini = mini_df.drop(['Listening_Time_minutes', 'id'], axis=1)
y= mini_df['Listening_Time_minutes']
```

```python
X_mini
```

```python
model = LinearRegression() 
model.fit(X_mini, y)
```

```python
import pickle

params = {
    "coef": model.coef_.tolist(),
    "intercept": model.intercept_.tolist()
}

with open("../Models/concept_test.pkl", "wb") as f:
    pickle.dump(params, f)
```
